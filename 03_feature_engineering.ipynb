```python
import pandas as pd

df = pd.read_csv('European_Bank.csv')
df.head()
```

```python
df['BalanceSalaryRatio'] = df['Balance'] / df['EstimatedSalary']
df[['Balance', 'EstimatedSalary', 'BalanceSalaryRatio']].head()
```

```python
df['HighProductDensity'] = (df['NumOfProducts'] >= 3).astype(int)
df[['NumOfProducts', 'HighProductDensity']].head(10)
```

```python
df['EngagementProductInteraction'] = df['IsActiveMember'] * df['NumOfProducts']
df[['IsActiveMember', 'NumOfProducts', 'EngagementProductInteraction']].head(10)
```

```python
df['AgeTenureInteraction'] = df['Age'] * df['Tenure']
df[['Age', 'Tenure', 'AgeTenureInteraction']].head(10)
```

```python
df.groupby('Exited')[['BalanceSalaryRatio', 'HighProductDensity', 'EngagementProductInteraction', 'AgeTenureInteraction']].mean()
```

```python
df.to_csv('European_Bank_features.csv', index=False)
```
