```python
import pandas as pd
import pickle
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingClassifier

# Load the feature-engineered data (before scaling/splitting)
df = pd.read_csv('European_Bank_features.csv')
df = df.drop(columns=['CustomerId', 'Surname', 'Year'])
df = pd.get_dummies(df, columns=['Geography', 'Gender'], drop_first=True)

X = df.drop(columns=['Exited'])
y = df['Exited']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Scale
numeric_cols = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 
                 'EstimatedSalary', 'BalanceSalaryRatio', 'EngagementProductInteraction', 
                 'AgeTenureInteraction']

scaler = StandardScaler()
X_train[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test[numeric_cols] = scaler.transform(X_test[numeric_cols])

# Train final model
gb_model = GradientBoostingClassifier(random_state=42, n_estimators=100, max_depth=3, learning_rate=0.1)
gb_model.fit(X_train, y_train)

print("Done training. Columns:", list(X.columns))
```

```python
with open('gb_model.pkl', 'wb') as f:
    pickle.dump(gb_model, f)

with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

with open('feature_columns.pkl', 'wb') as f:
    pickle.dump(list(X.columns), f)

print("All files saved!")
```
